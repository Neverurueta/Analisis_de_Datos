```python
plt.figure(figsize=(28,19))  ##tamaño para que quepan las tres imagenes
plt.subplot(221)  ### Dos filas 2 columnas
im=plt.imshow(b,cmap="Blues")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.subplot(222)
im=plt.imshow(r,cmap="Reds")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.subplot(223) #
im=plt.imshow(g,cmap="Greens")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.show()
```

```python
b1=b.ravel() ###convertimos la matriz del color azul en un vector para hacer el histograma
print(b1)
np.shape(b1)  ##el tamaño del vector debe ser igual a la multiplicación de las 
#filas por columnas de la matrizm b
r1=r.ravel()
g1=g.ravel()
```

```python
plt.figure(figsize=(8,6))  ##dimención de la imagen
plt.hist(b1,bins=256,range=(0,256),color='b',label='Azul', alpha=0.5)  ##color b grafica el color azul   label azul  lo identificamos en las leyendas
plt.hist(g1,bins=256,range=(0,256),color='g',label='green', alpha=0.5) # el alfa le da transparencia
plt.hist(r1,bins=256,range=(0,256),color='r',label='red', alpha=0.5)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))  ##dimención de la imagen
plt.hist(b1,bins=256,range=(0,256),color='b') 
plt.show()
```

```python
from scipy import stats 
```

La distribución de intensidad del color azul en la figura analizada es unimodal. La mayoría de las intensidades esta asociada a un valor de 71 (media ). El 50% de los pixeles son menores a 45 (mediana). Sin embargo considerando la desviaciónm estándar de 71, las inTensidades más probables, están en el intervalo de [0, 142]. la curtosis es 0 y el coeficiente de asimetría es 1,04 >0 significa que las intensidades más probables se alejan hacía la izquierda.

```python
print("La media es",np.round(np.mean(b1))) ##para acotar la media np.mean(b1[(g1>100)&(b1<170)])  Manera de ejemplo
print("La mediana es",np.round(np.median(b1)))
print("La desviación estandar es",np.round(np.std(b1)))
print("La media se mueve entre:",np.round(np.mean(b1)-np.std(b1)),"y",np.round(np.mean(b1)+np.std(b1)))
print("La curtosis es:",np.round(stats.kurtosis(b1)))
print("La moda es:",np.round(stats.mode(b1)))
print("El coeficiente de Asimetría es:",stats.skew(b1)) 
```
