```python
# PUNTO 4
#TONALIDAD DE LA ESCALA AZUL

Oscuros_Azul=len(b1[(b1>=0)&(b1<(255/3))])
Medio_Azul=len(b1[(b1>=(255/3))&(b1<((255/3)*2))])
Claro_Azul=len(b1[(b1>=((255/3)*2))&(b1<256)])
TA=Oscuros_Azul+Medio_Azul+Claro_Azul

AZUL= {'Tonalidad de la Escala Azul': ['Claro', 'Medio', "Oscuros"],
'Cantidad de Pixeles': [Claro_Azul, Medio_Azul, Oscuros_Azul],
'Porcentaje': [round(Claro_Azul*100/TA,2), round(Medio_Azul*100/TA,2), round(Oscuros_Azul*100/TA,2)]}
dfA = pd.DataFrame(AZUL,
columns=['Tonalidad de la Escala Azul', 'Cantidad de Pixeles', 'Porcentaje'])
dfA
```

```python
#TONALIDAD DE LA ESCALA VERDE
Oscuros_Verde=len(g1[(g1>=0)&(g1<(255/3))])
Medio_Verde=len(g1[(g1>=(255/3))&(g1<((255/3)*2))])
Claro_Verde=len(g1[(g1>=((255/3)*2))&(g1<256)])
TV=Oscuros_Verde+Medio_Verde+Claro_Verde

VERDE= {'Tonalidad de la Escala Verde': ['Claro', 'Medio', "Oscuros"],
'Cantidad de Pixeles': [Claro_Verde, Medio_Verde, Oscuros_Verde],
'Porcentaje': [round(Claro_Verde*100/TV,2), round(Medio_Verde*100/TV,2), round(Oscuros_Verde*100/TV,2)]}
dfV = pd.DataFrame(VERDE,
columns=['Tonalidad de la Escala Verde', 'Cantidad de Pixeles', 'Porcentaje'])
dfV
```

```python
#TONALIDAD DE LA ESCALA ROJA
Oscuros_Roja=len(r1[(r1>=0)&(r1<(255/3))])
Medio_Roja=len(r1[(r1>=(255/3))&(r1<((255/3)*2))])
Claro_Roja=len(r1[(r1>=((255/3)*2))&(r1<256)])
TR=Oscuros_Roja+Medio_Roja+Claro_Roja

ROJO= {'Tonalidad de la Escala Roja': ['Claro', 'Medio', "Oscuros"],
'Cantidad de Pixeles': [Claro_Roja, Medio_Roja, Oscuros_Roja],
'Porcentaje': [round(Claro_Roja*100/TR,2), round(Medio_Roja*100/TR,2), round(Oscuros_Roja*100/TR,2)]}
dfR = pd.DataFrame(ROJO,
columns=['Tonalidad de la Escala Roja', 'Cantidad de Pixeles', 'Porcentaje'])
dfR
```
